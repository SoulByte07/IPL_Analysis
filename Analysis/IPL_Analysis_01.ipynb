### import libraries

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

### üîç Dataset Overview

We are working with two datasets:

    matches.csv ‚Äì contains match-level data like season, teams, winner, toss decision, etc.

    deliveries.csv ‚Äì contains ball-by-ball delivery data (not deeply explored in this phase).

We‚Äôll perform data analysis and visualization using pandas, matplotlib, and seaborn.

```python
matches = pd.read_csv('../Datasets/Cleaned Datasets/cleaned_matches.csv')
deliveries = pd.read_csv('../Datasets/Cleaned Datasets/cleaned_deliveries.csv')

# print the first 5 rows of the loaded data
print(matches.head())
print(deliveries.head())
```

### üèÜ Top 5 Teams With Most Wins

Let‚Äôs begin by identifying the most successful teams in IPL history based on the total number of matches won.

```python
top_teams= matches['winner'].value_counts().head(5)

#plot 1
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5)) 
sns.barplot(x=top_teams.values,y=top_teams.index,hue=top_teams.index,palette='viridis',legend=False)
plt.title('top 5 teams(winners)')
plt.xlabel('no of wins')
plt.ylabel('teams')
plt.tight_layout()
plt.show()
plt.clf()
```

### Insight:
This plot shows the top 5 teams with the highest number of match wins in IPL. These teams have consistently performed well across seasons.

### üßç‚Äç‚ôÇÔ∏è Top 10 Players of the Match

Who are the most impactful players in the IPL? Let‚Äôs analyze the players who received the most "Player of the Match" awards.

```python
top_players = matches['player_of_match'].value_counts().head(10)

# Plot 2
sns.barplot(x=top_players.values, y=top_players.index, hue=top_players.index,palette="magma")
plt.title("Top 10 Players of the Match")
plt.xlabel("Number of Awards")
plt.ylabel("Player")
plt.show()
plt.clf()
```

### Insight:
These players have delivered outstanding performances in individual matches and have been game changers for their teams.

### üé≤ Does Winning the Toss Matter?

Let's analyze whether winning the toss increases the chances of winning the match.

```python
# Create a new column
matches['toss_match_win'] = matches['toss_winner'] == matches['winner']

# Count how often toss winner also won the match
toss_result = matches['toss_match_win'].value_counts()

# Plot 3
sns.barplot(x=toss_result.index, y=toss_result.values,hue=toss_result.index, palette="Set2")
plt.xticks([0, 1], ['Lost Match', 'Won Match'])
plt.title("Does Toss Winner Also Win Match?")
plt.ylabel("Number of Matches")
plt.show()
plt.clf()
```

### Insight:
From this chart, we observe whether toss winners have an actual advantage in match outcomes.

### üèüÔ∏è Most Popular Venues

Where are most IPL matches played? Let‚Äôs find out the top stadiums that host the maximum matches.

```python
venue_counts = matches['venue'].value_counts().head(10)

#plot 4
sns.barplot(y=venue_counts.index, x=venue_counts.values,hue=venue_counts.index, palette="cubehelix")
plt.title("Top 10 Most Used Venues")
plt.xlabel("No. of Matches")
plt.ylabel("Venue")
plt.show()
plt.clf()
```

### Insight:
These venues are IPL favorites, possibly due to capacity, facilities, or regional fanbase.

### üß¢ Toss Decision: Bat or Field?

What‚Äôs the usual choice after winning the toss‚Äîbat first or field first?

```python
#recult type counts
result_type=matches['toss_decision'].value_counts()

# Pie chart
plt.pie(result_type, labels=result_type.index, autopct='%1.1f%%', colors=['#FFDDC1', '#B5EAD7'])
plt.title("Match Wins: Batting First vs Chasing")
plt.show()
plt.clf()
```

### Insight:
The chart shows the common strategy teams follow after winning the toss‚Äîwhether they choose to set a target or chase.


### üìÖ Matches Played Per Season

Let‚Äôs see how the IPL has grown over time by checking the number of matches played each season.

```python
matches_per_year = matches['season'].value_counts().sort_index()

#plot 5
sns.lineplot(x=matches_per_year.index, y=matches_per_year.values, marker='o')
plt.title("IPL Matches Per Season")
plt.xlabel("Season")
plt.ylabel("Number of Matches")
plt.show()
plt.clf()
```

### Insight:
This visualization shows trends in how the tournament expanded or adjusted across different seasons.

### ü•á Most Successful Team Per Season

Let‚Äôs find the top team for each season based on the number of matches won.

```python
season_winner = matches.groupby(['season', 'winner']).size().reset_index(name='wins')

# Get max win team per season
top_per_season = season_winner.loc[season_winner.groupby('season')['wins'].idxmax()]

# Display as table
print(top_per_season[['season', 'winner', 'wins']])
```

### Insight:
This gives a season-by-season breakdown of the most dominant team. It's useful to understand shifts in team performance.
